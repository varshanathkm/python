1. triangle - check equilateral,isosceles,scalene

conditions

1. equilateral : 3 sides equal
2. scalene : 3 sides unequal
3. isosceles : 2 sides equal

```python
side1 = float(input("enter the 1st length"))
side2 = float(input("enter the 2nd length"))
side3 = float(input("enter the 3rd length"))

if side1 == side2 == side3:
  print("equilateral triangle")
elif side1 == side2 or side1 == side3 or side2 == side3:
  print("isosceles triangle")
else:
  print("scalene triangle")
```

1. accept two numbers
2. accept mathematical operator
3. print output

```python
a = int(input("enter 1st number"))
b = int(input("enter 2nd number"))
op = input("enter the operator")

if op == "+":
  print("your answer is:",a+b)
elif op == "-":
  print("your answer is:",a-b)
elif op == "*":
  print("your answer is:",a*b)
elif op == "/":
  print("your answer is:",a/b)
elif op == "%":
  print("your answer is:",a%b)
elif op == "//":
  print("your answer is:",a//b)
elif op == "**":
  print("your answer is:",a**b)
else:
  print("enter valid operator")
```
