#Difference method

the difference method returns the difference between the set and other sets , the difference is a set of elements that are in the first set but not in the other set

```python
data = {'raju','ramu','dolu','bolu','achu'}

data1 = {'kichu','michu','raju','ramu'}

result = data.difference(data1)

print(result)

```

#symmetric Difference

the symmetric difference method returns the symmetric difference of a set and another set , the symmetric difference is a set of elements that are in either of the set but not in both.

```python
data = {'raju','ramu','dolu','bolu','achu'}

data1 = {'kichu','michu','raju','ramu'}

result = data.symmetric_differencedifference(data1)

print(result)
```

#dict datatype

dictionary datatype is an unordered datatype, it is a mutable datatype , infinite length , it is iterable , it is a collection of key value pairs

```python
#syntax

#data = {'key':'value','key1':'value1'.......}
```

in this case keys are immutable datatype such as tuple , string , float, int etc

in values we can pass both mutable and immutable data

```python
data ={'raju':5000,'dolu':6000,'raju':7000,'bolu':9000}

print(data)
```

dictionary values are accessed by the keys of the value

```python
data ={'raju':5000,'dolu':6000,'raju':7000,'bolu':9000}

print(data)
print(data['raju'])
```

current key value can be updated

```python
data ={'raju':5000,'dolu':6000,'ramu':7000,'bolu':9000}

print(data)

data['raju'] = 9000

print(data)
```

new keyvalue can be added

```python
data ={'raju':5000,'dolu':6000,'ramu':7000,'bolu':9000}

print(data)

data['ammu'] = 9000

print(data)
```
