#conditional statements




*   it allows us to control the flow of execution based on condition that we specify
*   it is used to define the path of execution



```python
num = int(input('enter the number'))

if num%2!=0:

  print('number is odd')

else:

   print('number is even')
```

```python
num = int(input('enter the number'))

if num<0:

  print('number is negative')

else:

   print('number is positive')
```

```python
num = int(input('enter the number'))

if num<0:

  print('number is negative')

elif num ==0:

  print('number is zero')

else:

   print('number is positive')

print('completed')
```

##qlf = MBA exp >5 == snr hr
##qlf = MCA exp >5 == snr dev
##qlf = MCA == jnr dev
##qlf = BBA == intern hr
##qlf = BCA == inter dev

```python
qlf = input('enter the qlf')
exp = int(input('enter the experience'))

if qlf=='MBA' and exp>5:

  print('selected as senior hr')

elif qlf=='MCA' and exp>5:

  print('selected as senior dev')

elif qlf=='MCA':

  print('selected as junior dev')

elif qlf=='BBA':

  print('selected as intern hr')

elif qlf=='BCA':

  print('selected as intern dev')

else:

   print('your application doesnt meet the necessary criteria')
```
