#list creation

```python
data =['ramu','dolu','ramu','bolu','dolu']

print(data)
print(type(data))
```

#in built version

```python
data1 = list(('ramu','dolu', 'ramu', 'bolu', 'dolu'))
print(data1)
print(type(data1))
```

#nested list

```python
data = [['ramu',5000],['dolu',6000], 'ramu', 'bolu', 'dolu']
```

#list indexing and slicing

```python
data =['ramu','dolu','ramu','bolu','dolu']

#listname[index number]

print(data[0])
print(data[-1])
```

```python
data =['ramu',['dolu',56],'ramu','bolu','dolu']

#listname[index number]

print(data[0])
print(data[1])
```

```python
data =['ramu',['dolu',56],'ramu','bolu','dolu']

#listname[index number]
data1 = data[1]
print(data1)
print(data1[1])
```

#list element swapping

```python
data = ['ramu','achu','ramu','bolu','dolu']

data[1],data[3] = data[3],data[1]

print(data)
```

```python
data = ['ramu','achu','raju','bolu','dolu']

data[3],data[1],data[2],data[4],data[0] = data[0],data[3],data[4],data[1],data[2]

print(data)
```

```python
data = ['ramu','achu','raju','bolu','dolu']

data[2],data[3],data[0],data[1]=data[0],data[1],data[3],data[2]

print(data)
```

#list slicing

```python
data = ['raju','dolu','bolu','ramu','achu']

#listname[startindex:endvalue:stepvalue]

data1 = data[1:4]#end value given is with n+1 format to obtain the value that we need

print(data1)
```

```python
data = ['raju','dolu','bolu','ramu','achu']


data1 = data[0:4]

print(data1)
print(data[:4])
print(data[-5:-1])
print(data[-5:4])
print(data[:-1])
```

```python
data = ['raju','dolu','bolu','ramu','achu']

print(data[2:5])
print(data[-3:5])
print(data[2:])
print(data[-3:])
```

```python
data = ['raju','dolu','bolu','ramu','achu']

data[2]='ammu'

print(data)
```

#list methods

##append method

```python
data = ['raju','dolu','bolu','ramu','achu']

data.append('ammu')

print(data)
```

##extend method

```python
data = ['raju','dolu','bolu','ramu','achu']

data.extend(['ramu','achu','appu'])

print(data)
```

##insert method

```python
data = ['raju','dolu','bolu','ramu','achu']

data.insert(1,'ammu')

print(data)
```

##pop method

```python
data = ['raju','dolu','bolu','ramu','achu']

data.pop(1)

print(data)
```

```python
data = ['raju','dolu','bolu','ramu','achu']

data.pop()#automatically remove the last element if not specific value is given

print(data)
```

##remove method

```python
data = ['raju','achu','dolu','bolu','ramu','achu']

data.remove('achu')

print(data)
```

##clear method

```python
data = ['raju','achu','dolu','bolu','ramu','achu']

data.clear()

print(data)
```

##sort method

##ascending

```python
data = [23,22,5,1,8,6,56,45,32]

data.sort()

print(data)
```

##descending

```python
data = [23,22,5,1,8,6,56,45,32]

data.sort(reverse=True)

print(data)
```

##index method

```python
data = [23,22,5,1,8,6,56,45,32]

result=data.index(5)

print(result)
```

##copy method

```python
data = [23,22,5,1,8,6,56,45,32]

result=data.copy()

print(result)
```

##count method

```python
data = [23,22,5,1,8,6,56,45,32,5,5,5]

result=data.count(5)
print(result)
```

##reverse method

```python
data = [23,22,5,1,8,6,56,45,32,5,5,5]

data.reverse()

print(data)
```
