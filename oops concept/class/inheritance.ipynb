```python
class A:
  def printa(self):
    print("inside class A")

class B(A):
  def printb(self):
    print("inside class B")

obj = B()

obj.printb()
obj.printa()
```

```python
class student():

  def __init__(self,name,age,course):
    self.name=name
    self.age=age
    self.course=course

  def display(self):
    print(f"name: {self.name} age: {self.age} course: {self.course}")

class school(student):

  def __init__(self,name,age,course,school,place):
    student.__init__(self,name,age,course)
    self.school=school
    self.place = place

  def display_all(self):
    print(f"name: {self.name} age: {self.age} course: {self.course} school: {self.school} place: {self.place}")

obj = school('ben','16','ds','laliga','clt')
obj.display_all()
obj.display()
```

#multiple inheritance

```python
class A:

   def printa(self):
    print("inside class A")

class B:
  def printb(self):
    print("inside class B")

class C(A,B):
  def printc(self):
    print("inside class C")

obj = C()
obj.printc()
obj.printa()
obj.printb()
```

```python
class School:
    def __init__(self, school, place): # Corrected the constructor to __init__
        self.school = school
        self.place = place

    def display_school(self):
        print(self.school, self.place)


class College:
    def __init__(self, c_name, c_place): # Corrected the constructor to __init__
        self.c_name = c_name
        self.c_place = c_place

    def display_college(self):
        print(self.c_name, self.c_place)


class Students(School, College):
    def __init__(self, name, age, course, school, place, c_name, c_place): # Corrected the constructor to __init__
        School.__init__(self, school, place)
        College.__init__(self, c_name, c_place)
        self.name = name
        self.age = age
        self.course = course

    def display(self):
        print(self.name, self.age, self.course, self.school, self.place, self.c_name, self.c_place)

obj = Students('abhay', 23, 'DS', 'BLPS', 'MAVOOR', 'JCT', 'CSBS')
obj.display_school()
obj.display_college()
obj.display()
```

#multilevel inheritance

situation where a subclass inherits from another subclass

```python
class A:
  def printa(self):
    print("inside class A")

class B(A):
  def printb(self):
    print("inside class B")

class C(B):
  def printc(self):
    print("inside class C")

obj = C()
obj.printa()
obj.printb()
```
