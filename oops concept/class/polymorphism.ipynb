#method overloading

here only the second function gets called so providing only two inputs would give an error

```python
class A:

    def add(self,x,y):

      self.x = x
      self.y = y

      return self.x + self.y
# class B(A):
    def add(self,x,y,z=0):

      self.x = x
      self.y = y
      self.z = z

      return self.x + self.y + self.z

obj = A()

print(obj.add(10,20,30))
print(obj.add(10,20))
```

#method overriding

```python
class A:

    def add(self,x,y):

      self.x = x
      self.y = y

      return self.x + self.y
class B(A):
    def add(self,x,y,z):

      self.x = x
      self.y = y
      self.z = z

      return self.x + self.y + self.z

obj = B()

print(obj.add(10,20,30))
print(obj.add(10,20))
```

#encapsulation

wrapping up of data and methods . It is used to keep the data safe from interference and misuse.

it prevents accidental modification

```python
class students:

  def __init__(self,name,age,course):

    self.name = name
    self._age = age  #in case of protected(_) we can access or inherit but it will still be a protected data
    self.__course = course   # in case of private(__) we cant access any of the data

obj = students('sarang',21,'DS')

print(obj.name)
print(obj._age)
print(obj.__course)
```

```python
class person:

  def __init__(self,name,country,date_of_birth):

    self.name = name
    self.country = country
    self.date_of_birth = date_of_birth

  def age(self):

    current_year = 2024
    birth_year = int(self.date_of_birth[6:10])
    age = current_year - birth_year

    current_month = 11
    current_day = 4
    birth_month = int(self.date_of_birth[3:5])
    birth_day = int(self.date_of_birth[0:2])

    if current_month<birth_month or (current_month == birth_month and current_day<birth_day):

      age-=1

    return age

person1 = person('aju','india','01-12-2000')
print(person1.age())

```

```python
from datetime import datetime, date

class person:

  def __init__(self,name,country,date_of_birth):

    self.name = name
    self.country = country
    self.date_of_birth = date_of_birth

  def age(self):


    dob = datetime.strptime(self.date_of_birth, "%d-%m-%Y").date()  # strptime is used to interpret date string,.date(): This method is called on the resulting datetime object to extract only the date portion, creating a date object.


    today = date.today()


    age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day)) #If the birthday has already happened, this expression evaluates to False (which is equivalent to 0 in numerical context).If the birthday has not yet happened, the expression evaluates to True (which is equivalent to 1 in numerical context).

    return age

person1 = person('aju','india','01-12-2000')
print(person1.age())
```

2.write a python program to create a calculator class include methods for basic arithmetic operators

```python
class Calculator:
    def add(self, x, y):

        return x + y

    def subtract(self, x, y):

        return x - y

    def multiply(self, x, y):

        return x * y

    def divide(self, x, y):

        if y == 0:
            return "Cannot divide by zero"
        return x / y


calculator = Calculator()

print(calculator.add(5, 3))
print(calculator.subtract(10, 4))
print(calculator.multiply(2, 6))
print(calculator.divide(15, 5))
print(calculator.divide(15, 0))
```

3. create a dictionary named phonebook as class then using init method add data into the phonebook and using other methods update ,delete and remove data from the dictionary



```python
class Phonebook:
    def __init__(self,name,number):
        self.name = name
        self.number = number
        self.contacts = {}

    def __call__(self, name, number):
        self.contacts[name] = number
        print(f"Contact {name} added with number {number}.")


    def update_contact(self, name, new_number):

        if name in self.contacts:
            self.contacts[name] = new_number
            print(f"Contact {name} updated.")
        else:
            print(f"Contact {name} not found.")

    def delete_contact(self, name):

        if name in self.contacts:
            del self.contacts[name]
            print(f"Contact {name} deleted.")
        else:
            print(f"Contact {name} not found.")

    def remove_all_contacts(self):

        self.contacts.clear()
        print("All contacts removed.")

    def display_contacts(self):

        if self.contacts:
            print("Phonebook Contacts:")
            for name, number in self.contacts.items():
                print(f"{name}: {number}")
        else:
            print("Phonebook is empty.")


my_phonebook = Phonebook("John Doe", "123-456-7890")



my_phonebook.display_contacts()

# Update a contact
my_phonebook.update_contact("John Doe", "111-222-3333")

# Delete a contact
my_phonebook.delete_contact("Jane Smith")

# Display contacts again
my_phonebook.display_contacts()

# Remove all contacts
my_phonebook.remove_all_contacts()

# Display contacts again
my_phonebook.display_contacts()
```
