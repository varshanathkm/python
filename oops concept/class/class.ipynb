```python
class student():

  def details(self,name,age,course):
    self.name=name
    self.age=age
    self.course=course

  def display(self):

    print(self.name,self.age,self.course)
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Course: {self.course}")

obj = student()
obj.details('sarang','12','DS')
obj.display()
print('\r')
obj1 = student()
obj1.details('raju','23','testing')
obj1.display()
```

```python
import math
class circle():

  def radius(self,r):
    self.r=r
    self.area=math.pi*r*r
    self.circum=2*math.pi*r

  def display(self):
    print(f"Radius: {self.r}")
    print(f"Area: {self.area}")
    print(f"Circumference: {self.circum}")

obj = circle()
obj.radius(int('5'))
obj.display()

```

#init mthod.

it is used as a constructor. its mainly used to assign the values to data members. it is automatically called when an object is created

```python
import math
class circle():

  def __init__(self,r):
    self.r=r
  def area(self):
    return math.pi*self.r*self.r
  def perimeter(self):
    return 2*math.pi*self.r

obj = circle(5)
print(obj.perimeter())
print(obj.area())
```
