Definition: A database is an organized collection of data that is stored and managed to be easily accessed, updated, and analyzed.

Types of Databases:

1. Relational Databases (SQL): Use structured query language (SQL) and store data in tables. Examples include MySQL, PostgreSQL, and Oracle.

2. NoSQL Databases: Designed for unstructured or semi-structured data, focusing on flexibility and scalability. Examples include MongoDB, Redis, and Cassandra.

3. In-Memory Databases: Store data in RAM for faster access. Example: Redis.

4. Cloud Databases: Hosted on cloud platforms, providing scalability and remote access. Examples include Amazon RDS and Google BigQuery.

# Rules for naming a TABLE AND COLUMNS


* The name can contain letters (A-Z,a-z), digits(0-9), underscore(_) and dollar($) symbol.
* Name must contain at least one character. (names with only digits are invalid)
* Name must not contain white space and special symbols
* Name must not be an SQL keyword.
* The name should not duplicate with the names of other tables in the same database.
