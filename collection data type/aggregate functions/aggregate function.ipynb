the aggregate function in python are functions that perform a computation on a set of values to return a single result

#sum function

the sum function returns the sum of all elements in an iterable

```python
data = [23,22,5,1,8,6,56,45,32,5,5] #if string data exist in the list the sum wont be get

output = sum(data)

print(output)
```

##min function

returns the smallest element in an iterable

```python
data = [23,22,5,1,8,6,56,45,32,5,5]
output = min(data)

print(output)
```

##max function

used to return the largest element in an iterable

```python
data = [23,22,5,1,8,6,56,45,32,5,5]
output = max(data)

print(output)
```

#len function

returns the number of elements in an iterable

```python
data = [23,22,5,1,8,6,56,45,32,5,5]
output = len(data)

print(output)
```

#sorted function

##ascending

```python
data = [23,22,5,1,8,6,56,45,32,5,5]
output = sorted(data)

print(output)
```

##descending

```python
data = [23,22,5,1,8,6,56,45,32,5,5]
output = sorted(data,reverse=True)

print(output)
```
