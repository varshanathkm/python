#mode Description



1. "r" - Read - default value.open a file for reading,error if the file does not exists
2.   "w" - this mode opens file for writing. if file does not exist,it creates a new file. if the file exists it truncates the file.
3. "x" - creates a new file.if the file already exists, it creates a new file

4. "a" - open file in append mode.if file does not exists, it creates a new file
5. "t" - this is default mode. it opens in text mode

6. "b" - this opens in binary mode

7. "+" - this will open a file for reading and writing

create a file and write "file name 1 ,file name 2,.. upto 20" and then read the file

```python
f = open("demo2.txt",'w')
for i in range(1,21):
  f.write("this is line %d\r"%i)
f.close()

f= open("demo2.txt","r")
cont = f.read()

print(cont)
f.close()
```

```python
f = open("demo2.txt",'w')
for i in range(1,21):
  f.write("this is line"+str(i)+"\n")
f.close()

f= open("demo2.txt","r")
cont = f.read()

print(cont)
f.close()
```

#readline()

```python
f = open("demo2.txt","r")
cont = f.readline()
print(cont)

cont1 = f.readline()
print(cont1)
f.close()
```

file pointer position checking

```python
f = open("demo2.txt","r")

fp = f.tell()
print(fp)


cont = f.readline()
print(cont)

cont1 = f.readline()
print(cont1)
f.close()
```

#seek method

used to change the position of the file pointer

```python
f = open("demo2.txt","r")

fp = f.tell()
print(fp)


cont = f.readline()
print(cont)

f.seek(0)

cont1 = f.readline()
print(cont1)
f.close()
```

#append method

```python
f = open("demo2.txt","a")

f.write("this is append line")

f.close()

f= open("demo2.txt","r")
cont = f.read()

print(cont)
f.close()
```

read from one file and write to another file

```python
f = open("demo2.txt","r")
f1 = open("demo3.txt","w")

cont = f.read()
f1.write(cont)

f.close()
f1.close()

f1= open("demo3.txt","r")
cont = f1.read()

print(cont)
f1.close()
```
