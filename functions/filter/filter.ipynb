filter():
the filter function operator on a list and returns a subset of that of the list after applying the filter rule. The function filter(f,l) needs a function f as its arguments. f returns a boolean value ie, either true or false. this function will be applied to every element of the list l. only if f returns true will the element if the list l. only if f returns true will the element of the list be included in the result list.

```python
data = [1,5,2,5,2,7,5,6,4]

output = filter(lambda x: x%2==0,data)
print(list(output))
```

if the data has no vowels in it return the word into a new list

```python
data = ['raju','sarang','ramu','dolu','bolu','achu','nazu','vtyt']
def no_vowels(x):
    if 'a' in x or 'e' in x or 'i' in x or 'o' in x or 'u' in x:
        return False
    else:
        return True
output = filter(no_vowels,data)
print(list(output))
```

```python
data = ['raju','sarang','ramu','dolu','bolu','achu','nazu','vtyt']
vo = 'aeiouAEIOU'
def fun(x):
  for i in x:
    if i in vo:
      return False
  return True
print(list(filter(fun,data)))
```

```python
data = ['raju','sarang','ramu','dolu','bolu','achu','nazu','vtyt','p','q']
def no_vowels(x):
    if len(x)>1 and ('a' in x[1] or 'o' in x[1]):
        return True
    else:
        return False
output = filter(no_vowels,data)
print(list(output))
```

#map

The map() transforms a given list into a new list by transforming each element using a rule, it is used to apply a function to each element of a list of any other iterable

this is very useful when we do vector mathematics. the map() function will take multiple lists operates on them and return output list. one thing to note is that all the lists should be of the same size because map() applies the function  to corresponding elements in those lists

```python
# create a map () to create a square of number in side the list

data = [1,5,2,3,6,8]

out = lambda x:x**2
out = map(out,data)
print(list(out))


# or

def seq(x):
    return x**2
output = list(map(seq,data))
print(output)
```

```python
# calc the len of each element in the list and add to a list

data = ['abhay','ramu','achu']

def co(x):
    return len(x)

output = list(map(co,data))
print(output)
```

## **ZIP Function :**

*  zip() function creates an iterator that will aggregate elements from two or more iterables.

* syntax :               zip(*iterable)

```python
data = ['apple','banana','orange']
data1 = [1,5,6,8,4,2,5]

# out = list(zip(data,data1))
out = tuple(zip(data,data1))
# out = dict(zip(data,data1))
print(out)
```
