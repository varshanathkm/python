#decorator function

the decorator functions allows you to modify or extend the behaviour of functions or methods without changing their actual code, they are an excellent way to apply reusable functionality across multiple functions

```python
def sub(x,y):

  return x-y

print(sub(4,3))
print(sub(2,6))
```

```python
def div(x,y):
  p = x/y
  return p

print(div(4,2))
print(div(4,3))
print(div(2,6))
```

```python
def swap_num(fun):
  def wrapper(x,y):
    if x<y:
      x,y=y,x
    return fun(x,y)
  return wrapper

@swap_num
def sub(x,y):
  return x-y
@swap_num
def div(x,y):
  p = x/y
  return p

print(sub(4,3))
print(sub(2,6))
print(div(4,2))
print(div(4,1))
print(div(2,6))
```

#lambda functions

lambda functions are similar and user defined but without a name. They are commonly reffered to as anonymous functions.

syntax:

lambda argument(s): expression

lambda is a keyword in python for defining the anonymous function.
argument(s) is a placeholder, that is a variable that will be used to hold the value u want to pass into the function expression. A lambda function can have multiple variables depending on what u want to achieve.

expression is the code u want to execute in lambda function

```python
def div(a,b):
  return a/b
print(div(2,6))

x = lambda a,b:a/b

print(x(4,2))
```

```python
x = lambda a,b:a/b

print(x(4,2))
```
