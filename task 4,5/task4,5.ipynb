```python
num = (input("Enter a number: "))
sum = 0
p = int(num)
count = len(num)
while p > 0:
   digit = p % 10
   sum += digit ** count
   p //= 10
if int(num) == sum:
   print(num,"is an Armstrong number")
else:
   print(num,"is not an Armstrong number")
```

```python
num = int(input("Enter a number: "))

divisor_sum = 0
i = 1
while i < num:
    if num % i == 0:
        divisor_sum += i
    i += 1

if divisor_sum == num:
    print(num, "is a perfect number")
else:
    print(num, "is not a perfect number")
```
