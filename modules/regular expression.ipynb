## **match**

* match is used to Determine if the RE matches at the beginning of the string.

```python
# match

data ='hello iam kevin'

out = re.match('hello',data)        # match is used to Determine if the RE matches at the beginning of the string.
print(out)

# give condition so it will give the needed output only

out = re.match('hello',data)
if out:
    print('match')
else:
    print('not match')
```

## serach

* Scan through a string, looking for any location where this RE matches.

```python
# search

data ='hello iam kevin'

out = re.search('i',data)        # Scan through a string, looking for any location where this RE matches.
print(out)
```

#pwd valid
**length 6 to 16**

**atleast one lowercase**

**atleast one uppercase**

**atleast one digit**

**atleast one special character**

```python
def calculate_length(password):
    length = 0
    for i in password:
        length += 1
    return length

while True:
  y = input('enter password')
  password_length = calculate_length(y)
  if (password_length < 6 or password_length > 16):
    print('invalid password')
    print('length should be 6 to 16')

  elif not re.search('[a-z]',y):
    print('invalid password')
    print('atleast one lowercase')

  elif not re.search('[A-Z]',y):
    print('invalid password')
    print('atleast one uppercase')

  elif not re.search('[0-9]',y):
    print('invalid password')
    print('atleast one digit')

  elif not re.search('[^\w\s]',y):
    print('invalid password')
    print('atleast one special character')

  else:
    print('valid password')
    break
```
