```python
def add(a,b):
  return a+b #if given the b value as 10 then there is no need to specify args in the next code
```

```python
df['chem_marks']=df['chem_marks'].apply(add,args=(10,))
```

```python
df
```

```python
def add(x):

  if x== 'mahe':
    return 0
  elif x== 'kollam':
    return 1
  else:
    return 2
```

```python
df['place']=df['place'].apply(add)
```

```python
df['name']=df['name'].apply(lambda x:x.upper())
```

```python
df
```

```python
df1=df.groupby('place')

print(df1.count())
```

```python
df1.size()
```

```python
df1=df.groupby(['place','name'])

print(df1.count())
```
