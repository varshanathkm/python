```python
#merge- used to join the data in which we use a common column to merge it
#bydefault inner joiin is performed so we can specify by using how to outerjoin leftjoin etc
data= pd.merge(df,df1,on='MyUnknownColumn',how='outer')
```

```python
data= pd.merge(df,df1,on='MyUnknownColumn',how='left')
```

```python
data= pd.merge(df,df1,on='MyUnknownColumn',how='right')
```

#join method

```python
data=df.join(df1,lsuffix='_left',rsuffix='_right')#data in the left is specified with _left and data in rightside with _right so that no conflict occurs regarding same column name
```

#combine method

```python
import pandas as pd
```

```python
df=pd.DataFrame([[23,56,34],[23,56,34],[23,56,34]],columns=['a','b','c'])
df1=pd.DataFrame([[23,56,34],[23,56,34],[23,56,34]],columns=['a','b','c'])
```
