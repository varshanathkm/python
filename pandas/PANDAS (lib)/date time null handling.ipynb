#how to convert to timestamp

* %a weekdays as locales abbreviated name sun,mon
* %A weekdays as locales full name. sunday,monday
* %w weekdays as a decimal number monday:0,sunday:6
* %a weekday as locales abbreviated name sun,mon
* %d Day of month as a zero padded decimal number.01,02
* %b month as locales abbreviated name. jan,feb, dec
* %B month as locales full name. january
* %y year without centuary. 2000 as 00
* %Y year with centuary as a decimal number. 2013
* %H hour (24 hour clock). 00,01,...,23
* %I hour(12 hour clock). 00,01,...,12
* %M minute as zero padded decimal number.00,01,...,59
* %S second as a zero padded decimal number. 00,01,...,59

```python
df['year']=df['Timestamp'].dt.year
df['month']=df['Timestamp'].dt.month
df['day']=df['Timestamp'].dt.day
df['month_name']=df['Timestamp'].dt.month_name()
df['dayofweek']=df['Timestamp'].dt.dayofweek
df['day_name']=df['Timestamp'].dt.day_name()
df['dayofyear']=df['Timestamp'].dt.dayofyear
df['hour']=df['Timestamp'].dt.hour
df['minute']=df['Timestamp'].dt.minute
```
