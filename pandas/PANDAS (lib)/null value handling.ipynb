**NULL VALUE HANDLING**

no data area

```python
import pandas as pd
df = pd.read_csv('/content/sample_data/null_handling.csv',sep=',')
df
```

## **METHOD TO CHECK NULL VALUE**

```python
df.isnull()
```

```python
df.isna()
```

```python
# to check in which col have null value

df.isna().any()

# or

df.isnull().any()
```

```python
# to get the sum of null value in column

df.isna().sum()

#or

df.isnull().sum()
```

```python
# to check the % of null value in each columns

df.isnull().sum()*100/len(df)
```

## **TO DROP THE NULL VALUE IN THE COLUMN**

```python
df1=df.dropna()                   # to save this to actual data ( inplace = true)  , axis 1 = colum   , axis 0 = row
df1
```

```python
df.dropna(axis=1)                   #  , axis 1 = colum   , axis 0 = row
```

```python
 # to drop row with all value null

df.dropna(how='all')
```

```python
# to drop row which do not have min non-null values

df1 = df.dropna(thresh=3)
df1
```

## **METHODS TO FILL NULL VALUES**

```python
# add the value we given to null data

df.fillna(100)
```

```python
# to add data to null row (data from row above the null row)

# forward fill

df1 = df.fillna(method='ffill')
df1
```

```python
# to add data to null row (data from row bellow the null row)

# backward fill

df1=df1.fillna(method='bfill')
df1
```

* fill column with int or float value with **mean() or median()**

```python
df['marks'].mean()
```

```python
df['marks']= df['marks'].fillna(df['marks'].mean())
df
```

```python
df = pd.read_csv('/content/sample_data/null_handling.csv',sep=',')
df
```

```python
import numpy as np

df['marks']=df['marks'].replace({np.NaN:df['marks'].mean()})
df
```
