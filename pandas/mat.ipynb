#box plot

to visualize distribution of the data

also known as box-and-whisker plot

box plots components

median

First Quartile

Third Quartile

Lower Whisker

Upper whisker

Outliers

Steps to plot box plot

Median of data is calculated.

First quartile and third quartile is calculated.

IQR = thirdQuartile - firstQuartile

the IQR (Interquartile Range) is a measure of statistical dispersion. It's essentially the range of the middle 50% of the data.

Lower whisker = maximum(First quartile - 1.5*IQR )

Median: The median is the middle value of a dataset when it's sorted in ascending order. In a box plot, it's represented by the line inside the box.

First Quartile (Q1) and Third Quartile (Q3): These values divide the dataset into four equal parts. Q1 is the median of the lower half of the data, and Q3 is the median of the upper half.

Interquartile Range (IQR): It's the range between Q3 and Q1, calculated as IQR = Q3 - Q1. It represents the spread of the middle 50% of the data.

Lower Whisker: The lower whisker typically extends to the smallest data point within 1.5 times the IQR below the Q1. If there are no data points below this value, the lower whisker extends to the minimum value in the dataset.

Upper Whisker: The upper whisker extends to the largest data point within 1.5 times the IQR above the Q3. If there are no data points above this value, the upper whisker extends to the maximum value in the dataset.

Outliers: Data points beyond the whiskers are considered outliers and are plotted individually as points. They fall outside the range of the whiskers.

To create a box plot using Matplotlib:

Step 1: Calculate the median, Q1, and Q3 of your dataset.

Step 2: Compute the IQR using IQR = Q3 - Q1.

Step 3: Calculate the lower whisker as max(First quartile - 1.5 * IQR, minimum value in dataset).

Step 4: Calculate the upper whisker as min(Third quartile + 1.5 * IQR, maximum value in dataset).

#sub plot

useed to plot multiple graphs in single plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'horsepower':[120,160,145,200,170,190],
        'horseweight':[2500,3000,2700,2600,2300,2800],
        'horsespeed':[120,100,90,134,140,150]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
fig,axs=plt.subplots(3,3,figsize=(20,20))
axs[0,0].scatter(df['horsepower'],df['horseweight'],color='red',marker='*')
axs[0,0].set_xlabel('horsepower')
axs[0,0].set_ylabel('horseweight')
axs[0,0].set_title('scatter plot')
axs[0,0].grid(True,color='r')


axs[0,1].bar(df['horsepower'],df['horsespeed'],color='g',width=8)
axs[0,1].set_xlabel('horsepower')
axs[0,1].set_ylabel('horsespeed')
axs[0,1].set_title(' bar plot')
axs[0,1].grid(True,color='y')

axs[0,2].plot(df['horsepower'],df['horsespeed'],color='b',marker='^',markerfacecolor='y')
axs[0,2].set_xlabel('horsepower')
axs[0,2].set_ylabel('horsespeed')
axs[0,2].set_title(' line plot')
axs[0,2].grid(True,color='k')

axs[1,0].hist(df['horsepower'],color='orange',bins=5)
axs[1,0].set_xlabel('power')
axs[1,0].set_ylabel('frequency')
axs[1,0].set_title('histogram')
axs[1,0].grid(True,color='m')

axs[1,1].pie(df['horsepower'],labels=df.index,autopct='%1.2f%%')
axs[1,1].set_title('pie chart')

axs[1,2].boxplot(df['horsepower'],showmeans=True,patch_artist=True)
axs[1,2].set_title('boxplot')
axs[1,2].grid(True,color='c')

plt.style.use('dark_background')
plt.show()
```
