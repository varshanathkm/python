#loop control statements or jumping statements

the loop control statements are used to change the flow of execution, these can be used if u wish to skip an iteration or to stop the execution

```python
#continue - used to skip
#break
#pass
```

#continue statement - it is used within loops to skip the current iteration and proceed to the next one

instead of terminate the loop it forces to execute the next iteration of loop

```python
for i in range(5):

  print(i)

  continue
```

continue will skip the data

```python
for i in range(5):

  continue

  print(i)
```

```python
for i in range(10,21):

  if i%2!=0:

    continue

  print(i)
```

```python
employees=[
{"id":1000,"name":"hari", "dept":"devop","salary":34000},
{"id":1001, "name":"ravi", "dept":"devop","salary":24000},
{"id":1002, "name":"ram", "dept":"qa", "salary":37000},
{"id":1003, "name":"ajay", "dept":"ga","salary":14000},
{"id":1004, "name": "anu", "dept":"ba", "salary":54000},
{"id":1005, "name": "tini", "dept":"qa","salary":64000},]

for i in employees:
  if i["salary"]>25000:
    continue
  else:
    i.update({"salary":i["salary"]+5000})
    print(i)
```
