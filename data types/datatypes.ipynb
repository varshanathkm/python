##float to complex

```python
dat = 56.44

print(type(dat))
print(dat)

dat1=complex(dat)

print(type(dat1))
print(dat1)
```

```python
dat = 56.44

print(type(dat))
print(dat)

dat1=complex(dat,45)

print(type(dat1))
print(dat1)
```

in complex function we can declare int data but not a string data in the first value , we can still declare in the second value - this is only needed if we want to display the imaginary part

```python
data = 56

print(type(data))
print(data)

data1=complex(data,data)

print(type(data1))
print(data1)
```

we can also display the imaginary part directly

```python
data = 56

print(type(data))
print(data)

data1=complex(data,13)

print(type(data1))
print(data1)
```
