#exception

Events that interrupt the flow of a program during execution

#Exception Handling



*   exception handling allows you to seperate error handling code from normal code
*   Try and except statements are used to catch and handle exceptions in python.Statements that can raise exeptions are kept inside the try clause and the statements that handle the exception are written inside except clause



```python
#syntax

#Try :
   # code that may cause exception
#Except :
   # code to run when exception occurs
#Else:
   # code to run when exception does not occur
#finally :
   # code that will always be executed
```

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

try:
  output = num1/num2
  print(output)
except:
  print("no error")
finally:
  print("completed")
```

```python
try:
  num1 = int(input("Enter the first number: "))
  num2 = int(input("Enter the second number: "))
  output = num1/num2
  print(output)
except ZeroDivisionError as e:
  print(e)
except ValueError as e:
  print(e)
else:
  print("no error")
finally:
  print("completed")
```

```python
try:
  num1 = int(input("Enter the first number: "))
  num2 = int(input("Enter the second number: "))
  output = num1/num2
  print(output)
except Exception as e:
  print(e)
else:
  print("no error")
finally:
  print("completed")
```
