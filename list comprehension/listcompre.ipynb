#List comprehension

list comprehension offers a shorter syntax when u want to create a new list based on the values of a existing list

is a way to create a list by iterating over an iterable and optionally applying a condition

```python
#syntax

#data = [item for item in iterble conditional statements]
```

```python
x = range(20,40)

data = []

for i in x:
  if i % 2 == 0:
   data.append(i)
print(data)
```

```python
x = range(20,40)

data = [i for i in x if i % 2 == 0]
print(data)
```

if there is no conditional statement then all the numbers get appended to the list

```python
x = range(20,40)

data = [i for i in x]
print(data)
```
